```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd
import time
from IPython.display import clear_output, display  # Tambahan display untuk tabel

# URL untuk data gempa dirasakan oleh masyarakat
url = "https://data.bmkg.go.id/DataMKG/TEWS/gempadirasakan.xml"

# Fungsi untuk mengambil dan menampilkan data dalam tabel
def fetch_and_display_data():
    response = requests.get(url)
    data = response.content
    root = ET.fromstring(data)

    earthquake_data = []
    for gempa in root.findall("gempa"):
        magnitude = float(gempa.find("Magnitude").text)
        lokasi = gempa.find("Wilayah").text
        kedalaman = gempa.find("Kedalaman").text
        waktu = gempa.find("Tanggal").text + " " + gempa.find("Jam").text

        # Ambil Lintang & Bujur
        lintang = gempa.find("Lintang").text if gempa.find("Lintang") is not None else "Tidak Diketahui"
        bujur = gempa.find("Bujur").text if gempa.find("Bujur") is not None else "Tidak Diketahui"
        koordinat = f"{lintang}, {bujur}"

        earthquake_data.append([waktu, magnitude, lokasi, kedalaman, koordinat])

    # Konversi ke DataFrame
    df = pd.DataFrame(earthquake_data, columns=["Waktu", "Magnitudo", "Lokasi", "Kedalaman", "Koordinat"])

    # Bersihkan output lama & tampilkan data terbaru dalam bentuk tabel
    clear_output(wait=True)
    print(f"ðŸ“¢ Data diperbarui pada: {time.strftime('%Y-%m-%d %H:%M:%S')}")

    # Tampilkan tabel interaktif
    display(df)

# Loop untuk refresh otomatis setiap 2 menit
while True:
    fetch_and_display_data()
    time.sleep(120)  # Tunggu 2 menit sebelum refresh data
```
